```python
# IMPORT LIBRARY
import pandas as pd
import numpy as np
import keras
import tensorflow as tf
from keras.layers import Dense
from keras import Sequential
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# IMPORT DATASET
dataset = pd.read_csv("Dataset/tinggi_badan.csv",sep=";")
dataset
```

```python
# INISIASI INPUT & OUTPUT
x = dataset.loc[:,["tinggi_ayah","berat_ayah","tinggi_ibu","berat_ibu","gender"]]
y = dataset["tinggi_anak"]
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
x["gender"] = le.fit_transform(x["gender"])
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
column_sels = ["tinggi_ayah","berat_ayah","tinggi_ibu","berat_ibu","gender"]
x = x.loc[:,column_sels]
y = dataset['tinggi_anak']
x = pd.DataFrame(data=min_max_scaler.fit_transform(x), columns=column_sels)
```

```python
x = x.values
y = y.values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=32)
```

```python
X_train
```

```python
# # INISIASI ANN (Artificial Neural Network)
# classifier = Sequential()

# # INISIASI INPUT LAYER DAN HIDDEN LAYER SEBANYAK 5 NODE
# classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu', input_dim = 5))

# # PENDAMBAHAN HIDDEN LAYER KEDUA 
# classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu'))

# # PENDAMBAHAN HIDDEN LAYER KETIGA
# classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu'))

# # PENDAMBAHAN HIDDEN LAYER KELIMA
# classifier.add(Dense(units = 15, kernel_initializer = 'uniform', activation = 'relu'))

# # PENAMBAHAN OUTPUT LAYER
# classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'relu'))

regressor = keras.Sequential([
keras.layers.Dense(15, activation=tf.nn.relu, input_shape=[5]),
keras.layers.Dense(12, activation=tf.nn.relu),
keras.layers.Dense(13, activation=tf.nn.relu),
keras.layers.Dense(1, activation=tf.nn.relu)
])

regressor.compile(optimizer='adam',loss='mean_absolute_error')
save_best_val_loss = keras.callbacks.ModelCheckpoint('best_model_val_loss.h5',save_best_only=True,monitor='val_loss',mode='min')
save_best_loss = keras.callbacks.ModelCheckpoint('best_model_loss.h5',save_best_only=True,monitor='loss',mode='min')
```

```python
history = regressor.fit(X_train,y_train,validation_data=(X_test, y_test),epochs = 10,batch_size = 400,shuffle = True,callbacks = [save_best_val_loss, save_best_loss])
```
